```python
import numpy as np


def replace_empty(val):
    if isinstance(val, (list, np.ndarray)) and len(val) == 0:
        return [["del"]]
    return val

# Step 2: Replace [["none"]] with [["normal"]]
def replace_none(val):
    if isinstance(val, (list, np.ndarray)) and np.array_equal(val, [["none"]]):
        return [["normal"]]
    return val

# Apply to relevant columns only
df["failureType"] = df["failureType"].apply(replace_empty).apply(replace_none)
df["trianTestLabel"] = df["trianTestLabel"].apply(replace_empty)
```

```python
import os
import random

# Define source and destination paths
source_dir = r"C:\Users\anshi\Downloads\wafer_defect\wafer_images_last_100000"
train_dir = r"C:\Users\anshi\Downloads\wafer_defect\wafer_split\train"
test_dir = r"C:\Users\anshi\Downloads\wafer_defect\wafer_split\test"

# Make sure train and test folders exist
os.makedirs(train_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

# Collect all image paths with labels
all_images = []

for label in os.listdir(source_dir):
    class_path = os.path.join(source_dir, label)
    if not os.path.isdir(class_path):
        continue

    for image_name in os.listdir(class_path):
        image_path = os.path.join(class_path, image_name)
        all_images.append((image_path, label))

# Sort by filename or modified time if needed (optional)
all_images = sorted(all_images, key=lambda x: x[0])  # by filename
# all_images = sorted(all_images, key=lambda x: os.path.getmtime(x[0]))  # by timestamp

# Select last 100000 images
last_100k_images = all_images[-100000:]

# Group again by class
split_data = {}

for img_path, label in last_100k_images:
    if label not in split_data:
        split_data[label] = []
    split_data[label].append(img_path)

# Final dictionary to hold train/test
final_split = {}

for label, image_paths in split_data.items():
    random.shuffle(image_paths)
    split_index = int(0.8 * len(image_paths))
    final_split[label] = {
        'train_images': image_paths[:split_index],
        'test_images': image_paths[split_index:]
    }

print("✅ Last 100,000 image paths prepared and split into train/test.")
```

```python
import shutil
import os

# Copy the images to train/test directories using final_split
for label, data in final_split.items():
    # Create label folders inside train and test dirs
    train_label_dir = os.path.join(train_dir, label)
    test_label_dir = os.path.join(test_dir, label)
    os.makedirs(train_label_dir, exist_ok=True)
    os.makedirs(test_label_dir, exist_ok=True)

    # Copy training images
    for src in data['train_images']:
        dst = os.path.join(train_label_dir, os.path.basename(src))
        shutil.copy(src, dst)

    # Copy testing images
    for src in data['test_images']:
        dst = os.path.join(test_label_dir, os.path.basename(src))
        shutil.copy(src, dst)

print("✅ All last 100,000 images copied into respective train/test folders.")
```

```python
import os
import matplotlib.pyplot as plt
df['failureType'] = df['failureType'].apply(
    lambda x: x[0][0] if isinstance(x, list) and len(x) > 0 and isinstance(x[0], list) and len(x[0]) > 0 else x
)
df['trianTestLabel'] = df['trianTestLabel'].apply(
    lambda x: x[0][0] if isinstance(x, list) and len(x) > 0 and isinstance(x[0], list) and len(x[0]) > 0 else x
)


# Count images in each class folder
def count_images_per_class(base_dir):
    class_counts = {}
    for class_name in os.listdir(base_dir):
        class_path = os.path.join(base_dir, class_name)
        if os.path.isdir(class_path):
            image_count = len([
                fname for fname in os.listdir(class_path)
                if fname.lower().endswith(('.png', '.jpg', '.jpeg'))
            ])
            class_counts[class_name] = image_count
    return class_counts

# Directories
train_dir = r"C:\Users\anshi\Downloads\wafer_defect\wafer_split\train"
test_dir = r"C:\Users\anshi\Downloads\wafer_defect\wafer_split\test"

# Get counts
train_counts = count_images_per_class(train_dir)
test_counts = count_images_per_class(test_dir)

# Combine all class names
all_classes = sorted(set(train_counts.keys()) | set(test_counts.keys()))

# Prepare bar data
train_values = [train_counts.get(cls, 0) for cls in all_classes]
test_values = [test_counts.get(cls, 0) for cls in all_classes]

# Plotting
x = range(len(all_classes))
bar_width = 0.35

plt.figure(figsize=(14, 6))
bars1 = plt.bar(x, train_values, width=bar_width, label='Train', color='skyblue')
bars2 = plt.bar([i + bar_width for i in x], test_values, width=bar_width, label='Test', color='salmon')

# Annotate train values
for bar in bars1:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 5, str(height), ha='center', va='bottom', fontsize=9)

# Annotate test values
for bar in bars2:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 5, str(height), ha='center', va='bottom', fontsize=9)

plt.xlabel('Class Labels')
plt.ylabel('Number of Images')
plt.title('Image Count per Class in Train and Test Sets')
plt.xticks([i + bar_width / 2 for i in x], all_classes, rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Set image shape
IMG_HEIGHT = 64
IMG_WIDTH = 64
BATCH_SIZE = 32

# Define paths (use raw strings to avoid backslash issues)
train_dir = r"C:\Users\anshi\Downloads\wafer_defect\wafer_split\train"
test_dir = r"C:\Users\anshi\Downloads\wafer_defect\wafer_split\test"

# Create ImageDataGenerators
train_datagen = ImageDataGenerator(rescale=1.0 / 255)
test_datagen = ImageDataGenerator(rescale=1.0 / 255)

# Create data generators
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='categorical'  # Use 'categorical' for multi-class classification
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='categorical'
)

source_dir = r"C:\Users\anshi\Downloads\wafer_defect\wafer_images_last_100000"
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

# Number of classes from train_generator
num_classes = len(train_generator.class_indices)

# ✅ Define CNN model
model = Sequential([
    # Convolution Layer 1
    Conv2D(32, (3, 3), activation='relu', input_shape=(IMG_HEIGHT, IMG_WIDTH, 3)),
    MaxPooling2D(pool_size=(2, 2)),

    # Convolution Layer 2
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),

    # Flatten and Dense layers
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.3),
    Dense(num_classes, activation='softmax')  # Final layer
])
model
```

```python
model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)


model.summary()
```

```python
EPOCHS = 20  # You can increase later

history = model.fit(
    train_generator,
    epochs=EPOCHS,
    validation_data=test_generator
)
```

```python
import matplotlib.pyplot as plt

# Accuracy Plot
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()

# Loss Plot
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```

```python
test_loss, accuracy = model.evaluate(test_generator)
print(f"Test Loss: {test_loss:.4f}")   
print(f"Test Accuracy: {accuracy:.4f}")

train_loss, train_accuracy = model.evaluate(train_generator)
print(f"Train Loss: {train_loss:.4f}")
print(f"Train Accuracy: {train_accuracy:.4f}")
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns

# Get predictions
Y_pred = model.predict(test_generator)
y_pred = np.argmax(Y_pred, axis=1)

# True labels
y_true = test_generator.classes

# Confusion Matrix
cm = confusion_matrix(y_true, y_pred)
labels = list(test_generator.class_indices.keys())

# Plot
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Optional: Full classification report
print(classification_report(y_true, y_pred, target_names=labels))
```
